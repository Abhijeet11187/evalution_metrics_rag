```python
# Build the RAG pipeline
```

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
rag_prompt="""
You are an assitant who is expert in question-answering tasks.
Answer the following question using only the following pieces of retrieved context.
If the answer is not in the context, do not makeup the answer,just say that you dont know.
Keep the answer to the point based on the information from the context.

Question:
{question}
Context:
{context}

Answer:
"""
```

```python
rag_prompt_template=ChatPromptTemplate.from_template(rag_prompt)
```

```python
rag_prompt_template.input_variables
```

```python
## LLM Model (Langchain model for building RAG system)
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda
from operator import itemgetter
```

```python
chatgpt=ChatOpenAI(model_name="gpt-4o-mini",temperature=0)
```

```python
def format_docs(docs):
  return "\n\n".join(doc.page_content for doc in docs)
```

```python
src_rag_response_chain=(
    {
        "context":(itemgetter("context"))|RunnableLambda(format_docs),
        "question":RunnablePassthrough()
    }
    |rag_prompt_template
    |chatgpt
    |StrOutputParser()
)
```

```python
rag_chain_with_src_data=(
    {
        "context":similarity_retriever,
        "question":RunnablePassthrough()
    }
    |RunnablePassthrough.assign(response=src_rag_response_chain)
)
```

```python
# 4) LLM as a judge (G-Eval) -COT to evaluate LLM output based on any custom criteria
```

```python
from deepeval.test_case import LLMTestCase
from deepeval.metrics import GEval
from deepeval import evaluate
from deepeval.test_case import LLMTestCaseParams
```

```python
query="What is AI"
result=rag_chain_with_src_data.invoke(query)

result
```

```python
ai_response=result['response']
```

```python
ai_response
```

```python
human_answer="""
AI, also known as Artificial Intelligence is used to build complex systems for applications like virtual assistants,
robotics, and autonomous vehicles.
"""
```

```python
retrieved_context=[doc.page_content for doc in result['context']]
```

```python
test_case=LLMTestCase(
    input=result['question'],
    actual_output=ai_response,
    expected_output=human_answer,
    retrieval_context=retrieved_context
)
```

```python
metrics=GEval(
    threshold=0.5,
    model="gpt-4o",
    name="RAG Fact Checker",
# Note : You can provide either criteria or evalution steps ,and not both
    evaluation_steps=[
        "Create a list of statement from 'actual output'",
        "Validate if they are relevant and answers the given question in 'input',penalize if any statements are irrelevant",
        "Also validate if they esist in 'expected output',penalize if any statements are missing or factually wrong",
        "Also validate if these statements are grounded in the 'retrieval context' and penalize if they are missing or factually incorrect",
        "Finally also penalize of any statement seem to be invented or made up and do not make sense factually given the 'input' and 'retrieval context'"
    ],
    evaluation_params=[
        LLMTestCaseParams.INPUT,
        LLMTestCaseParams.ACTUAL_OUTPUT,
        LLMTestCaseParams.EXPECTED_OUTPUT,
        LLMTestCaseParams.RETRIEVAL_CONTEXT
    ],
    verbose_mode=True

)
```

```python
metrics_result=evaluate([test_case],[metrics])
```

```python
print("Success",metrics_result.test_results[0].metrics_data[0].success)
print("Score",metrics_result.test_results[0].metrics_data[0].score)
print("Reason",metrics_result.test_results[0].metrics_data[0].reason)
```
